````python
import matplotlib.pyplot as plt
````

````python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# plt.plot(x, y) 선으로 그래프 그려지는거
# plt.bar(x, y) 막대그래프 그려지는거
# plt.barh(x, y) 수평 막대그래프 그려지는거
plt.bar(x, y)
plt.show()
````

````python
s = [1, 2, 3, 4, 5]
score = [60,65,70,85,100]
# plt.scatter(s, score) 점으로 그래프 그려지는거
plt.bar(s, score)
plt.xlabel('study time')
plt.ylabel('score')
plt.title('study time vs score')
plt.show()
````
